<b> Importing library</b>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reading data set

data = pd.read_csv("http://bit.ly/w-data")

# data sample
data.head(5)
```

#Preprocessing data

```python
#data shape
print("data shape is: ",data.shape)
```

```python
#checking data for null value

data.isnull().sum()
```

```python
#data visualization

sns.set(style='whitegrid')
sns.regplot(data = data, x='Hours', y='Scores' , color='blue')
```

```python
# set the target data as y
X = data.iloc[:, :-1].values

y = data.iloc[:, 1].values
```

```python
#splitting data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

#Training the model

```python
#import linearRegression  model

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

#fit the data into linearRegression model
result =regressor.fit(X_train, y_train)

```

```python
print(X_test )
#predict the result of test data

y_pred = regressor.predict(X_test)
```

```python
#convert it into dataFrame 
df = pd.DataFrame({'Actural': y_test, 'Predicted': y_pred})
df
```

#Predict the error in module

```python
from sklearn import metrics
print("Mean absolute error: ", metrics.mean_absolute_error(y_test, y_pred))
```
